```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the data
df = pd.read_csv("datasets/office_episodes.csv")

df.head()
```

```python
df.info()
```

```python
#update guest_stars as string
df.guest_stars = df.guest_stars.astype('str')
```

```python
df.guest_stars[0] == "nan"
```

```python
# get the interested features from df
target = df[['episode_number','viewership_mil','scaled_ratings']]
target['marker_size'] = df.guest_stars.apply(lambda x: 25 if x == "nan" else 250)

target.head()
```

```python
#get the color of each point based on scaled ratings
colors = []

for i, row in target.iterrows():
    if row.scaled_ratings < 0.25:
        colors.append("red")
    elif row.scaled_ratings >= 0.25 and row.scaled_ratings < 0.5:
        colors.append("orange")
    elif row.scaled_ratings >= 0.5 and row.scaled_ratings < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
        
#plot scatter, episode number vs viewership
fig = plt.figure(figsize=(12,6))
#plt.rcParams['figure.figsize'] = [11, 7]
plt.scatter(target.episode_number, target.viewership_mil, c=colors, s=target.marker_size)
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.grid()
plt.show()
```

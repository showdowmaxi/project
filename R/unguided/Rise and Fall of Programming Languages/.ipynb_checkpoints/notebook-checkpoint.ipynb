## Questions:
1. What fraction of the total number of questions asked in 2019 had the R tag? 
2. What were the five most asked-about tags in the last 5 years (2015-2020)?

```r
library(ggplot2)
library(dplyr)
library(readr)
```

```r
stack_overflow <- read_csv("datasets/stack_overflow_data.csv")
head(stack_overflow)
```

### 1. What fraction of the total number of questions asked in 2019 had the R tag?

```r
r <- stack_overflow%>%filter(year == 2019, tag =='r') %>%mutate(percent = (number/year_total)*100)
r_percentage <- r$percent
r_percentage
```

### 2. What were the five most asked-about tags in the last 5 years (2015-2020)?

```r
#get the past 5 years data from 2015- 2020 top 5 tag 
past_5_yr <- stack_overflow %>% filter(year %in% 2015:2020) %>% 
        group_by(tag) %>% 
        summarise(tag_total = sum(number)) %>% 
        arrange(desc(tag_total))
highest_tags <- head(past_5_yr,5)$tag
highest_tags
```

```r
#plot the change of these top 5 tags  from 2015 - 2020
top_5 <-stack_overflow %>% filter(year %in% 2015:2020, tag %in% highest_tags)%>%mutate(fraction = number/year_total)

ggplot(top_5, aes(year, fraction, color = tag))+
    geom_line()+
    labs(title="Top 5 tags from 2015 - 2020 in Stack Overflow")
```

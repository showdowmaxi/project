```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
```

```python
df = pd.read_csv('datasets/logins.csv')
df.head()
```

```python
#check length
df['pw_len'] = df.password.apply(lambda x:len(x) >= 10)
df.head()
```

```python
#one lower case
df['lower_case'] = df.password.str.contains(r'[a-z]')
df.head()
```

```python
#one upper case
df['upper_case'] = df.password.str.contains(r'[A-Z]')
df.head()
```

```python
#one digits
df['digits'] = df.password.str.contains(r'\d')
df.head()
```

```python
#non-alphanumeric character
df['non_alphanumeric'] = df.password.str.contains(r'\W')
df.head()
```

```python
#no contains phrass password 
df['no_contain_password'] = ~df.username.str.lower().str.contains('password')
df.head()
```

```python
# contains first name and last name
df['first_name'] = df.username.apply(lambda x:x.split('.')[0])
df['last_name'] = df.username.apply(lambda x:x.split('.')[1])
df.head()
```

```python
#contains name

#set the contain_name as False 
df['no_contain_name'] = True

for i, row in df.iterrows():
    if row.first_name in row.password.lower() or row.last_name in row.password.lower():
        # set contain_name as True
        df.at[i, 'no_contain_name'] = False
df.head()
```

```python
#get the bad_pw df
bad_pw = df[~(df.pw_len & df.lower_case & df.upper_case & df.digits & df.non_alphanumeric & df.no_contain_name & df.no_contain_password)]

#calculate bad_pass
bad_pass = round(len(bad_pw)/len(df), 2)
bad_pass
```

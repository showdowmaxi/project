```python
# import relevant libraries
import nltk
from bs4 import BeautifulSoup
import requests
from collections import Counter
```

```python
# get the content from the source
soup = BeautifulSoup(requests.get("https://www.gutenberg.org/files/16/16-h/16-h.htm").text)

text = ""
#extract all the content in paragaph
for t in soup.find_all('p'):
    text = text + t.getText()

#print first 20 words 
print(text[:20])
```

```python
#lowercase the text
content = text.lower()

#print first 20 words 
print(content[:20])
```

```python
#keep all words 
tokenizer = nltk.tokenize.RegexpTokenizer('[a-zA-Z]\w+')
words = tokenizer.tokenize(content)
print(words[:20])
```

```python
#remove all stop words from english
stopwords = nltk.corpus.stopwords.words('english')
words_ns = [w for w in words if w not in stopwords]
```

```python
#get the most 10 common words
print(Counter(words_ns).most_common(10))
```
